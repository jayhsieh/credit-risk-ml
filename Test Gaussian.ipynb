```python
from sklearn import datasets
iris = datasets.load_iris()
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
import pandas as pd
# read data into a DataFrame
data1 = pd.DataFrame.from_csv('data1.csv', sep=',', index_col=0)
data2 = pd.DataFrame.from_csv('data2.csv', sep=',', index_col=0)
data3 = pd.DataFrame.from_csv('data3.csv', sep=',', index_col=0)
data4 = pd.DataFrame.from_csv('data4.csv', sep=',', index_col=0)
data5 = pd.DataFrame.from_csv('data5.csv', sep=',', index_col=0)
test = pd.DataFrame.from_csv('test.csv', sep=',', index_col=0)

data1.feature_account_type.replace(['checking', 'savings'], [1, 0], inplace=True)
data2.feature_account_type.replace(['checking', 'savings'], [1, 0], inplace=True)
data3.feature_account_type.replace(['checking', 'savings'], [1, 0], inplace=True)
data4.feature_account_type.replace(['checking', 'savings'], [1, 0], inplace=True)
data5.feature_account_type.replace(['checking', 'savings'], [1, 0], inplace=True)
test.feature_account_type.replace(['checking', 'savings'], [1, 0], inplace=True)

data1.label_error_code.replace(['No Error', 'ACH_INSUFFICIENT_FUNDS'], [0, 1], inplace=True)
data2.label_error_code.replace(['No Error', 'ACH_INSUFFICIENT_FUNDS'], [0, 1], inplace=True)     
data3.label_error_code.replace(['No Error', 'ACH_INSUFFICIENT_FUNDS'], [0, 1], inplace=True)     
data4.label_error_code.replace(['No Error', 'ACH_INSUFFICIENT_FUNDS'], [0, 1], inplace=True)     
data5.label_error_code.replace(['No Error', 'ACH_INSUFFICIENT_FUNDS'], [0, 1], inplace=True)     
test.label_error_code.replace(['No Error', 'ACH_INSUFFICIENT_FUNDS'], [0, 1], inplace=True) 

data1X = data1.filter(['feature_transaction_usd_send', 'feature_local_receive','feature_kount_score',
                    'feature_kount_velocity', 'feature_sift_score',
                    'feature_six_month_send_limit', 'feature_account_type', 
                     'feature_customer_age', 'feature_previous_transactions'])
data1Y = data1.filter(['label_error_code'])


data2X = data2.filter(['feature_transaction_usd_send', 'feature_local_receive','feature_kount_score',
                    'feature_kount_velocity', 'feature_sift_score',
                    'feature_six_month_send_limit', 'feature_account_type', 
                     'feature_customer_age', 'feature_previous_transactions'])
data2Y = data2.filter(['label_error_code'])


data3X = data3.filter(['feature_transaction_usd_send', 'feature_local_receive','feature_kount_score',
                    'feature_kount_velocity', 'feature_sift_score',
                    'feature_six_month_send_limit', 'feature_account_type', 
                     'feature_customer_age', 'feature_previous_transactions'])
data3Y = data3.filter(['label_error_code'])


data4X = data4.filter(['feature_transaction_usd_send', 'feature_local_receive','feature_kount_score',
                    'feature_kount_velocity', 'feature_sift_score',
                    'feature_six_month_send_limit', 'feature_account_type', 
                     'feature_customer_age', 'feature_previous_transactions'])
data4Y = data4.filter(['label_error_code'])


data5X = data5.filter(['feature_transaction_usd_send', 'feature_local_receive','feature_kount_score',
                    'feature_kount_velocity', 'feature_sift_score',
                    'feature_six_month_send_limit', 'feature_account_type', 
                     'feature_customer_age', 'feature_previous_transactions'])
data5Y = data5.filter(['label_error_code'])

testX = test.filter(['feature_transaction_usd_send', 
                            'feature_local_receive',
                            'feature_kount_score',
                            'feature_kount_velocity',
                            'feature_sift_score',
                            'feature_six_month_send_limit',
                            'feature_account_type',
                            'feature_customer_age',
                            'feature_previous_transactions'])
testY = test.filter(['label_error_code'])


#X = data1X
X = data2X
#X = X.append(data2X)
X = X.append(data3X)
X = X.append(data4X)
X = X.append(data5X)

#Y = data1Y
Y = data2Y
#Y = Y.append(data2Y)
Y = Y.append(data3Y)
Y = Y.append(data4Y)
Y = Y.append(data5Y)

testX = data1X
testY = data1Y

y_pred = gnb.fit(X, Y['label_error_code']).predict(testX)
print("Number of mislabeled points out of a total %d points : %d accuracy : %0.2f" % 
      ( testX.shape[0],
        (testY['label_error_code'] != y_pred).sum(),
        (((testX.shape[0]-(testY['label_error_code'] != y_pred).sum())*100)/testX.shape[0])
        #(testY['label_error_code'] != y_pred and y_pred == 'No Error').sum(),
        #(testY['label_error_code'] != y_pred and y_pred == 'ACH_INSUFFICIENT_FUNDS').sum()
      )
)

fp = 0;
fn = 0;
for actual, predicted in zip(testY['label_error_code'], y_pred):
    #print("actual %s predicted %s" % (actual, predicted))
    if(actual != predicted and actual == 0):
            fp+=1
    if(actual != predicted and actual == 1):
            fn+=1
        
print("False Positive (Predicted as Insufficient Funds when it is No Error) : %d" % (fp))
print("False Negative (Predicted as No Error when it is Insufficient Funds) : %d" % (fn))

```

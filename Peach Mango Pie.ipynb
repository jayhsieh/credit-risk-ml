```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification
```

```python
data = pd.read_csv('nov-last-5-ach (4).csv', sep=',', index_col=0)
data.head()
```

```python
data.columns
```

```python
X = data.filter(['feature_transaction_usd_send', 'feature_local_receive', 'feature_kount_score'])
X.head()
```

```python
data.label_error_code.replace(['No Error', 'ACH_INSUFFICIENT_FUNDS'], [1, 0], inplace=True)
y = data.filter(['label_error_code'])
y.head()
```

```python
#X, y = make_classification(n_samples=1000, n_features=4,
#                           n_informative=2, n_redundant=0,
#                           random_state=0, shuffle=False)
clf = RandomForestClassifier(max_depth=2, random_state=0)
```

```python
clf.fit(X, y.values.ravel())
```

```python
print(clf.feature_importances_)
```

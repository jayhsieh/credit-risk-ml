```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('nov-last-5-ach (4).csv', sep=',', index_col=0)
data.head()
```

```python
data.columns
```

```python
data.feature_account_type.replace(['checking', 'savings'], [1, 0], inplace=True)
X = data.filter(['feature_transaction_usd_send', 'feature_local_receive', 
                 'feature_kount_score', 'feature_kount_velocity', 'feature_sift_score',
       #'feature_days_since_first_completed', 
                 'feature_six_month_send_limit',
       'feature_account_type'
                 #, 'feature_day_of_month'
                 ,'feature_customer_age'
                 , 'feature_previous_transactions'
                ])
X.head()
```

```python
data.label_error_code.replace(['No Error', 'ACH_INSUFFICIENT_FUNDS'], [1, 0], inplace=True)
y = data.filter(['label_error_code'])
y.head()
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
#X, y = make_classification(n_samples=1000, n_features=4,
#                           n_informative=2, n_redundant=0,
#                           random_state=0, shuffle=False)
clf = RandomForestClassifier(max_depth=7, random_state=0)
```

```python
clf.fit(X_train, y_train.values.ravel())
```

```python
print(clf.feature_importances_)
```

```python
y_pred = clf.predict(X_test)
```

```python
unique, counts = np.unique(y_pred, return_counts=True)
dict(zip(unique, counts))
```

```python
print("Number of mislabeled points out of a total %d points : %d accuracy %0.2f" % 
      ( X_test.shape[0],
        (y_test['label_error_code'] != y_pred).sum(),
        ((X_test.shape[0]-(y_test['label_error_code'] != y_pred).sum())*100)/X_test.shape[0]
      )
     )
```

```python
unique, counts = np.unique(y_train, return_counts=True)
dict(zip(unique, counts))
```

```python
clf.score(X_test, y_test)
```

```python
y_pred_prob = clf.predict_proba(X_test)
```

```python
unique, counts = np.unique(y_pred_prob, return_counts=True)
dict(zip(unique, counts))
```
